```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================================================================
# 1. LOAD & PREPARE HEALTHCARE ADMINISTRATIVE DATASETS
# ==============================================================================
# Load patient demographics and service encounters datasets
patient_demographics = pd.read_csv('QVI_purchase_behaviour.csv')
service_encounters = pd.read_csv('QVI_transaction_data.csv')

# Standardize column headers for healthcare operations modeling
patient_demographics.rename(columns={
    'LYLTY_CARD_NBR': 'PATIENT_ID',
    'LIFESTAGE': 'DEMOGRAPHIC_GROUP',
    'PREMIUM_CUSTOMER': 'COVERAGE_TIER'
}, inplace=True)

service_encounters.rename(columns={
    'DATE': 'ENCOUNTER_DATE',
    'STORE_NBR': 'FACILITY_ID',
    'LYLTY_CARD_NBR': 'PATIENT_ID',
    'TXN_ID': 'ENCOUNTER_ID',
    'PROD_NBR': 'SERVICE_CODE',
    'PROD_NAME': 'SERVICE_DESCRIPTION',
    'PROD_QTY': 'UNITS_PROVIDED',
    'TOT_SALES': 'BILLED_AMOUNT'
}, inplace=True)

# Data Hygiene: Remove administrative duplicates
service_encounters = service_encounters.drop_duplicates()

# Convert Excel serial dates/integers to Datetime
service_encounters['ENCOUNTER_DATE'] = pd.to_datetime(
    service_encounters['ENCOUNTER_DATE'], 
    origin='1899-12-30', 
    unit='D'
)

# Remove non-operational outlier records
service_encounters = service_encounters[service_encounters['UNITS_PROVIDED'] < 100]

# Merge into unified Healthcare Operations Dataset
df_health_ops = pd.merge(service_encounters, patient_demographics, on='PATIENT_ID', how='left')

# ==============================================================================
# 2. FEATURE ENGINEERING & SERVICE CATEGORIZATION
# ==============================================================================
# Extract unit volume specifications using regex
df_health_ops['SERVICE_UNIT_SPEC'] = df_health_ops['SERVICE_DESCRIPTION'].str.extract('(\d+)').astype(float)

# Extract primary service category keyword
df_health_ops['SERVICE_CATEGORY'] = df_health_ops['SERVICE_DESCRIPTION'].str.split().str[0]

# ==============================================================================
# 3. HIGH-UTILIZER & RESOURCE CAPACITY ANALYSIS
# ==============================================================================
# Aggregate utilization metrics per patient
patient_utilization = df_health_ops.groupby('PATIENT_ID').agg(
    total_visits=('ENCOUNTER_ID', 'count'),
    total_billed=('BILLED_AMOUNT', 'sum'),
    avg_encounter_cost=('BILLED_AMOUNT', 'mean')
).reset_index()

# Define High-Utilizers (Top 10% highest visit frequency)
utilization_threshold = patient_utilization['total_visits'].quantile(0.90)

patient_utilization['UTILIZATION_TIER'] = np.where(
    patient_utilization['total_visits'] >= utilization_threshold,
    'High Utilizer (Top 10%)',
    'Standard Utilizer'
)

# Merge utilization tier back to primary dataset
df_health_ops = pd.merge(df_health_ops, patient_utilization[['PATIENT_ID', 'UTILIZATION_TIER']], on='PATIENT_ID', how='left')

# ==============================================================================
# 4. EXECUTIVE OPERATIONS REPORTING & VISUALIZATION
# ==============================================================================
plt.style.use('seaborn-v0_8-whitegrid' if 'seaborn-v0_8-whitegrid' in plt.style.available else 'default')

# Visualization: Healthcare Expenditure by Demographic & Coverage Tier
plt.figure(figsize=(11, 6))
demo_summary = df_health_ops.groupby(['DEMOGRAPHIC_GROUP', 'COVERAGE_TIER'])['BILLED_AMOUNT'].sum().unstack()
demo_summary.plot(kind='barh', stacked=True, figsize=(11, 6), colormap='Blues')
plt.title('Total Healthcare Service Expenditure by Demographic & Coverage Tier', fontsize=13, fontweight='bold')
plt.xlabel('Total Billed Expenditure ($)')
plt.ylabel('Patient Demographic Group')
plt.tight_layout()
plt.savefig('healthcare_expenditure_by_demographic.png')
plt.show()

# Summary Metrics
print("=" * 60)
print("         HEALTHCARE OPERATIONS ANALYTICS SUMMARY         ")
print("=" * 60)
print(f"Total Patient Encounters Processed: {len(df_health_ops):,}")
print(f"Total Billed Service Revenue:      ${df_health_ops['BILLED_AMOUNT'].sum():,.2f}")
print(f"Unique Patient Cohort Size:        {df_health_ops['PATIENT_ID'].nunique():,}")
print(f"High-Utilizer Visit Threshold:     >= {int(utilization_threshold)} visits")
print("=" * 60)
```
